```python
test_df = pd.read_csv('/kaggle/input/ieee-fraud-detection/test_transaction.csv')
test_identity = pd.read_csv('/kaggle/input/ieee-fraud-detection/test_identity.csv')
X_test = test_df.merge(test_identity, on='TransactionID', how='left')
transaction_ids = X_test['TransactionID']
X_test = X_test.drop(columns=['TransactionID']).copy()
```

```python
dagshub.init(repo_owner='gioeba', repo_name='IEEE-CIS-Fraud-Detection', mlflow=True)
```

```python
logged_model = "models:/xgboost_prod/1"
loaded_pipeline = mlflow.sklearn.load_model(logged_model)

y_test_pred = loaded_pipeline.predict_proba(X_test)[:, 1]

submission = pd.DataFrame({
    'TransactionID': transaction_ids,
    'isFraud': y_test_pred
})

submission.to_csv('submission.csv', index=False)
print("Submission saved")
```
